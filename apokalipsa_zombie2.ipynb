# Apokalipsa zombie

Apokalipsa zombie od lat fascynuje zarówno twórców popkultury, jak i naukowców, którzy próbują modelować niezwykłe scenariusze przetrwania ludzkości. Wizje świata opanowanego przez nieumarłych, choć wydają się abstrakcyjne, oferują wyjątkową możliwość eksploracji fundamentalnych problemów matematycznych, biologicznych i społecznych. Projekt ten stawia przed sobą ambitne zadanie: wykorzystać matematyczne modele do zrozumienia, przewidywania i potencjalnego odwrócenia skutków hipotetycznej pandemii zombie.

Koncept apokalipsy zombie wywodzi się z legend i tradycji ludowych, szczególnie z haitańskiego voodoo, gdzie zombie byli przedstawiani jako ożywione zwłoki kontrolowane przez magiczne siły. Współczesny obraz zombie, jako wolno poruszających się drapieżników, został spopularyzowany dzięki filmowi Night of the Living Dead George'a Romero z 1968 roku. Od tego czasu zombie stały się stałym elementem popkultury, jednocześnie inspirując naukowców do badań nad ich hipotetycznym wpływem na społeczeństwo.

**Motywacje i cel projektu**

Na pierwszy rzut oka badania nad apokalipsą zombie mogą wydawać się groteskowe lub nierealne. Jednak w rzeczywistości takie modele pozwalają na rozwój narzędzi analitycznych, które mogą być wykorzystane w innych dziedzinach, takich jak epidemiologia, ekologia czy logistyka kryzysowa. W przypadku rzeczywistej pandemii, jak COVID-19, szybkie reakcje i efektywne strategie zarządzania zależały od wcześniejszych symulacji i matematycznych predykcji. Zombie, jako koncepcyjni "nosiciele infekcji", stanowią modelowy przykład analiz, które mogą rzucić nowe światło na dynamikę rozprzestrzeniania się chorób oraz interakcji między populacjami.
Celem niniejszego projektu jest opracowanie realistycznych strategii przetrwania, bazując na zaawansowanych modelach matematycznych. Przeanalizujemy kluczowe aspekty, takie jak:
Czas, jaki zajmuje horda zombie, by dotrzeć do linii obrony ludzi.
Możliwości spowolnienia infekcji w grupach mieszanych (ludzie i zombie).
Skuteczność działań obronnych oraz wpływ geografii na wynik walki.
Z perspektywy naukowej, projekt ten pozwala przełamać schematyczne podejście do dynamiki epidemii. Poprzez włączenie wymiaru przestrzennego i interakcji społecznych, zyskujemy bardziej kompleksowy obraz inwazji. Takie podejście jest kluczowe w rzeczywistych scenariuszach kryzysowych, gdzie infekcje, dostęp do zasobów i mobilność populacji odgrywają decydującą rolę.

**Struktura** 

W ramach tego projektu przedstawione zostaną kluczowe wyniki badań matematycznych oraz strategie pozwalające ludzkości na przetrwanie w hipotetycznym scenariuszu. Zaczniemy od omówienia podstawowego modelu infekcji zombie, następnie rozwiniemy go o wymiar przestrzenny oraz dodatkowe interakcje między populacjami ludzi i zombie. Na końcu przeanalizujemy uzyskane dane i opracujemy konkretne strategie działania.
To właśnie matematyka, nauka uznawana za „język wszechświata”, stanie się naszym kluczem do odkrycia, jak ludzkość może odroczyć, a być może nawet uniknąć swojego końca w obliczu nieumarłych przeciwników.


### Definicje
**Apokalipsa** – katastroficzne wydarzenie prowadzące do zniszczenia dotychczasowego porządku świata. W kontekście omawianego problemu apokalipsy zombie następuje przejęcie terytoriów przez nowe, destrukcyjne siły, takie jak omawiane zombie, które eliminują lub zmieniają dotychczasowych mieszkańców, rozszerzając swoje panowanie.

**Zombie** - martwa lub ożywiona istota pozbawiona świadomości, która wykazuje agresję i instynktowną chęć polowania na ludzi. Jej celem jest nie tylko pożeranie ludzi, ale również przemiana ich w zombie poprzez zakażenie lub ukąszenie, co prowadzi do rozprzestrzeniania się ich gatunku.

# Model inwazji zombie z grupą zainfekowanych
$$
\begin{cases}
S' = -\beta SZ - \delta S, \\
I' = \beta SZ - (\rho + \delta)I, \\
Z' = \rho I - \alpha SZ + \zeta R, \\
R' = \delta(S + I) + \alpha SZ - \zeta R.
\end{cases}
$$

gdzie:
* $S$ - liczba ludzi niezarażonych, ale podatnych
* $Z$ - liczba aktywnych zarażających zombie
* $R$ - liczba zmarłych
* $I$ - liczba zainfekowanych (podlegają oni stopniowej przemianie, w trakcie której nie są jeszcze zagrożeniem dla innych)
* $\alpha$ - współczynnik, który odpowiada za prawdopodobieństwo tego, że zombie zginie
* $\beta$ - współczynnik, który odpowiada za prawdopodobieństwo tego, że w wyniku takiego spotkania człowiek zostanie zarażony
* $\delta$ - parametr decydujący o umieralności
* $\rho$ -  parametr odpowiadający za prędkość przemiany.
* $\zeta$ - parametr decydujący o przemianie zmarłych w zombie

### Punkty równowagi
**Punkt równowagi** jest to taki stan układu dynamicznego, w którym układ pozostaje w spoczynku, jeśli znajdzie się w tym punkcie. Matematycznie oznacza to, że wszystkie zmienne opisujące dynamikę układu pozostają stałe w czasie, ponieważ siły lub procesy zmieniające te zmienne równoważą się.

Teraz zajmiemy się wyznaczeniem punktów równowagi powyższego układu równań. 
Zgodnie z definicją, podstawimy wszystkie równania do $0$
$$
\begin{cases}
-\beta SZ - \delta S = 0, \\
\beta SZ - (\rho + \delta)I = 0, \\
\rho I - \alpha SZ + \zeta R = 0, \\
\delta(S + I) + \alpha SZ - \zeta R = 0.
\end{cases}
$$

Z pierwszego równania wynika, że $Z = - \frac{\delta}{\beta}$ lub $S = 0$. 

Podstawiając $Z = - \frac{\delta}{\beta}$ otrzymujemy: $S\in \mathbb{R}$, $I = - \frac{\delta S}{\rho + \delta}$, $R = \frac{S}{\zeta}(\frac{\rho \delta}{\rho + \delta} - \frac{\alpha \delta}{\beta})$.

Stąd pierwszy punkt równowagi wynosi $P_1 = (S, - \frac{\delta S}{\rho + \delta}, \frac{\delta}{\beta}, \frac{S}{\zeta}(\frac{\rho \delta}{\rho + \delta} - \frac{\alpha \delta}{\beta}))$, gdzie $S\in \mathbb{R}$. Ten scenariusz jednak **nie ma sensu fizycznego**, ponieważ $Z < 0$, co oznacza **brak żyjących zombie**.

Podstawiając teraz $S = 0$ do pozostałych równań, otrzymujemy: $I = 0$, $R = 0$, natomiast z ostatniego równania wynika, iż $Z\in \mathbb{R}$. 

Stąd drugi punkt równowagi wynosi $P_2 = (0,0,Z,0)$, gdzie $Z\in \mathbb{R}$.

Badamy następnie stabilność drugiego punktu równowagi. Wyznaczamy macierz Jakbiego, która opisuje aproksymację układu w otoczeniu punktów równowagi.

$$
J(S,I,Z,R)= 
\begin{pmatrix}
\frac{\partial S'}{\partial S} & \frac{\partial S'}{\partial I} & \frac{\partial S'}{\partial Z} & \frac{\partial S'}{\partial R} \\
\frac{\partial I'}{\partial S} & \frac{\partial I'}{\partial I} & \frac{\partial I'}{\partial Z} & \frac{\partial I'}{\partial R} \\
\frac{\partial Z'}{\partial S} & \frac{\partial Z'}{\partial I} & \frac{\partial Z'}{\partial Z} & \frac{\partial Z'}{\partial R} \\
\frac{\partial R'}{\partial S} & \frac{\partial R'}{\partial I} & \frac{\partial R'}{\partial Z} & \frac{\partial R'}{\partial R}
\end{pmatrix}
=
\begin{pmatrix}
-\beta Z - \delta - \lambda & 0 & -\beta S & 0 \\
\beta Z & -(\rho + \delta + \lambda) & \beta S & 0\\
-\alpha Z & \rho & -\alpha S & \zeta \\
\rho + \alpha Z & \delta & 0& -\zeta
\end{pmatrix}
$$
Podstawiamy punkt $P_2$:
$$
J(P_2) = 
\begin{pmatrix}
-\beta Z - \delta & 0 & 0 & 0 \\
\beta Z & -(\rho + \delta) & 0 & 0\\
-\alpha Z & \rho & 0 & \zeta \\
\rho + \alpha Z & \delta & 0 & -\zeta
\end{pmatrix}
$$

Poszukujemy pierwiastków wielomianu charakterystycznego, by wyznaczyć wartości własne:
$$
\text{det} (J(P_2) - \lambda I) = 
\begin{pmatrix}
-\beta Z - \delta - \lambda & 0 & 0 & 0 \\
\beta Z & -(\rho + \delta + \lambda) & 0 & 0\\
-\alpha Z & \rho & -\lambda & \zeta \\
\rho + \alpha Z & \delta & 0 & -\zeta - \lambda
\end{pmatrix}
$$

Stosujemy rozwinięcie Laplace'a względem trzeciej kolumny:
$$
\text{det} (J(P_2) - \lambda I) = (-\lambda)\text{det}
\begin{pmatrix}
-\beta Z - \delta - \lambda & 0 & 0 \\
\beta Z & -(\rho + \delta + \lambda) & 0\\
\rho + \alpha Z & \delta & -\zeta - \lambda
\end{pmatrix}
$$

Teraz stosujemy rozwinięcie Laplace'a również względem trzeciej kolumny:
$$
\text{det} (J(P_2) - \lambda I) = -\lambda(-\zeta - \lambda)
\begin{pmatrix} 
-\beta Z - \delta - \lambda & 0 \\
\beta Z & -(\rho + \delta + \lambda)
\end{pmatrix}
=
-\lambda(-\zeta - \lambda)(-\beta Z - \delta - \lambda)(-\rho - \delta - \lambda) = 0
$$

Wyznaczone wartości własne to
$$
\lambda _1 = 0, \lambda _2 = - \zeta, \lambda _3 = -\beta Z - \delta, \lambda _4 = -\rho - \delta.
$$

### Wnioski
Nie dysponujemy odpowiednimi narzędziami analitycznymi pozwalającymi na samodzielne rozwiązanie powyższego układu równań. W związku z tym, w celu przedstawienia rozwiązania, powołujemy się na wyniki przedstawione w artykule
["WHEN ZOMBIES ATTACK!: MATHEMATICAL MODELLING OF AN OUTBREAK OF ZOMBIE INFECTION"](https://math.uchicago.edu/~shmuel/Modeling/WHEN%20ZOMBIES%20ATTACK!-%20MATHEMATICAL%20MODELLING%20OF%20AN%20OUTBREAK%20OF%20ZOMBIE%20INFECTION.pdf). Analiza wskazuje, że omawiany punkt równowagi jest **stabilny**, co oznacza, że w tym scenariuszu zombie ostatecznie doprowadzą do **pełnej infekcji całej populacji**. Innymi słowy, niezależnie od początkowych warunków, proces rozprzestrzeniania się infekcji zmierza w kierunku całkowitej dominacji zombie nad zdrowymi osobnikami.

# Model inwazji zombie z możliwością wyleczenia
$$
\begin{cases}
S' &= -\beta SZ - \delta S + cZ, \\
I' &= \beta SZ - (\rho + \delta)I, \\
Z' &= \rho I - \alpha SZ - cZ + \zeta R, \\
R' &= \delta(S + I) + \alpha SZ - \zeta R.
\end{cases}
$$

gdzie:
* $c$ - parametr decydujący o powrocie wyleczonych zombie do bycia człowiekiem (jednak są one wciąż podatne na ponowne zarażenie)

### Punkty równowagi

W podobny sposób wyznaczamy punkty krytyczne dla drugiego modelu, dla prostoty przyjmujemy $\delta = 0$. (Zakładając krótki czas trwania epidemii wpływ współczynnika umieralności na wyniki nie będzie duży).

$$
\begin{cases}
-\beta SZ +cZ = 0, \\
\beta SZ - \rho I = 0, \\
\rho I - \alpha SZ - cZ +\zeta R = 0, \\
\delta (S+I) + \alpha SZ -\zeta R = 0.
\end{cases}
$$

Z pierwszego równania wynika, że $Z=0$ lub $S=\frac{c}{\beta}$. 

Dla $Z=0$ otrzymujemy: $S\in \mathbb{R}$,  $I=0$ i $R = 0$.

Stąd pierwszy punkt równowagi wynosi $P_1=(S,0,0,0)$, gdzie $S$ jest dowolne. Odpowiada on za **brak epidemii**, gdyż wszystkie grupy poza niezarażonymi są zerowe.

Podstawiając teraz $S=\frac{c}{\beta}$ otrzymujemy: $I = \frac{c}{\rho}Z$, $Z \in \mathbb{R}$ i $R = \frac{\alpha c}{\beta \zeta}Z$.

Stąd drugi punkt równowagi wynosi $P_2=(\frac{c}{\beta},\frac{c}{\rho}Z,Z,\frac{\alpha c}{\beta \zeta}Z)$, gdzie $Z$ jest dowolne.

Tak jak dla poprzedniego modelu, do sprawdzenia stabilności punktów równowagi wykorzystujemy macierz Jakobiego.

$$
J(S,I,Z,R) = 
\begin{pmatrix}
\frac{\partial S'}{\partial S} & \frac{\partial S'}{\partial I} & \frac{\partial S'}{\partial Z} & \frac{\partial S'}{\partial R} \\
\frac{\partial I'}{\partial S} & \frac{\partial I'}{\partial I} & \frac{\partial I'}{\partial Z} & \frac{\partial I'}{\partial R} \\
\frac{\partial Z'}{\partial S} & \frac{\partial Z'}{\partial I} & \frac{\partial Z'}{\partial Z} & \frac{\partial Z'}{\partial R} \\
\frac{\partial R'}{\partial S} & \frac{\partial R'}{\partial I} & \frac{\partial R'}{\partial Z} & \frac{\partial R'}{\partial R} \\
\end{pmatrix}
=
\begin{pmatrix}
-\beta Z & 0 & -\beta S + c& 0 \\
\beta Z & -\rho & \beta S & 0 \\
-\alpha Z & \rho & -\alpha S - c & \zeta \\
\alpha Z & 0 & \alpha S & - \zeta \\
\end{pmatrix}
$$
Aby wyznaczyć wartości własne, szukamy pierwiastków wielomianu charakerystycznego:

$$
\text{det} (J(P_2)-\lambda I) = \text{det}
\begin{pmatrix}
-\beta Z -\lambda & 0 & 0 & 0 \\
\beta Z & -\rho -\lambda& c & 0 \\
-\alpha Z & \rho & -\frac{\alpha c}{\beta} - c -\lambda & \zeta \\
\alpha Z & 0 & \frac{\alpha c}{\beta}  & - \zeta -\lambda \\
\end{pmatrix}
$$

Stosujemy rozwinięcie Laplace'a względem 1. wiersza:

$$
\text{det} (J(P_2)-\lambda I) = (-\beta Z - \lambda)\text{det}
\begin{pmatrix}
-\rho -\lambda& c & 0 \\
\rho & -\frac{\alpha c}{\beta} - c -\lambda & \zeta \\
0 & \frac{\alpha c}{\beta}  & - \zeta -\lambda \\
\end{pmatrix}
\stackrel{w_2+w_1} {=} (-\beta Z - \lambda)\text{det}
\begin{pmatrix}
-\rho -\lambda& c & 0 \\
-\lambda & -\frac{\alpha c}{\beta} -\lambda & \zeta \\
0 & \frac{\alpha c}{\beta}  & - \zeta -\lambda \\
\end{pmatrix} \\
\stackrel{w_2+w_3} {=}
(-\beta Z - \lambda)\text{det}
\begin{pmatrix}
-\rho -\lambda& c & 0 \\
-\lambda & -\lambda & -\lambda \\
0 & \frac{\alpha c}{\beta}  & - \zeta -\lambda \\
\end{pmatrix}
\stackrel{k_2-k_1} {=}
(-\beta Z - \lambda)\text{det}
\begin{pmatrix}
-\rho -\lambda& c + \rho + \lambda & 0 \\
-\lambda & 0 & -\lambda \\
0 & \frac{\alpha c}{\beta}  & - \zeta -\lambda \\
\end{pmatrix} \\
\stackrel{k_3-k_1} {=}
(-\beta Z - \lambda)\text{det}
\begin{pmatrix}
-\rho -\lambda& c + \rho + \lambda & \rho + \lambda \\
-\lambda & 0 & 0 \\
0 & \frac{\alpha c}{\beta}  & - \zeta -\lambda \\
\end{pmatrix} 
$$

Stosujemy rozwinięcie Laplace'a względem 2. wiersza:

$$
\text{det} (J(P_2)-\lambda I) = \lambda (\lambda +\beta Z) \text{det}
\begin{pmatrix}
- c + \rho + \lambda & \rho + \lambda \\
 \frac{\alpha c}{\beta}  & - \zeta -\lambda \\
\end{pmatrix}
= -\lambda (\lambda +\beta Z)\left(\lambda ^2 + \lambda\left(c+\rho +\zeta +\frac{\alpha c}{\beta}\right)+c\zeta +\rho \zeta + \frac{\rho \alpha c}{\beta}\right) \\
= -\lambda(\lambda +\beta Z)\left(\lambda - \frac{-c-\rho -\zeta -\frac{\alpha c}{\beta}+\sqrt{\Delta}}{2}\right)\left(\lambda - \frac{-c-\rho -\zeta -\frac{\alpha c}{\beta}-\sqrt{\Delta}}{2}\right) \text{, gdzie}\quad \Delta = \left(c+\rho +\zeta +\frac{\alpha c}{\beta}\right)^2 - 4\left(c\zeta +\rho \zeta + \frac{\rho \alpha c}{\beta}\right)
$$

Wyznaczone wartości własne to:
$$
\lambda _1 = 0, \quad \lambda _2 = -\beta Z, \quad \lambda _3 = \frac{-c-\rho -\zeta -\frac{\alpha c}{\beta}+\sqrt{\Delta}}{2}, \quad \lambda _4 = \frac{-c-\rho -\zeta -\frac{\alpha c}{\beta}-\sqrt{\Delta}}{2}
$$

Tak jak w poprzednim modelu, z rozumowania przedstawionego w artykule ["WHEN ZOMBIES ATTACK!: MATHEMATICAL MODELLING OF AN OUTBREAK OF ZOMBIE INFECTION"](https://math.uchicago.edu/~shmuel/Modeling/WHEN%20ZOMBIES%20ATTACK!-%20MATHEMATICAL%20MODELLING%20OF%20AN%20OUTBREAK%20OF%20ZOMBIE%20INFECTION.pdf) wynika, że punkt $P_2$ jest **stabilny**. Oznacza to, że zombie nie doprowadzą do całkowitej zagłady ludzkości. Układ będzie dążył do równowagi, w której będą współistnieć zarówno ludzie jak i zombie.

# Metody numeryczne
Niech $$\frac{dy}{dx}=y^{\prime}=f(x,y)$$
- **Metoda Eulera** jest jedną z najprostszych metod numerycznych stosowanych do rozwiązywania równań różniczkowych zwyczajnych (ODE). Zakłada, że zmiana funkcji na małym przedziale czasowym $h$ (gdzie $h$ to krok czasowy) jest liniowa, czyli nachylenie funkcji $f(x,y)$ jest stałe na tym przedziale. Na tej podstawie, używając wartości funkcji i jej pochodnej w punkcie początkowym, metoda Eulera przybliża wartość funkcji w kolejnym punkcie. 
$$y_{i+1}​=y_i​+hf(x_i​,y_i​), \qquad y(x_0) = y_0$$
- **Metoda Heuna** jest metodą dwuetapową, czyli przy obliczaniu wartości funkcji w punkcie $x_{i+1}$​ uwzględnia zarówno informacje z punktu $x_i$ ​, jak i szacunkowe wartości w punkcie x_{i+1}​. Ta metoda uwzględnienia średnią wartość nachylenia na przedziale $[x_i,x_{i+1}]$. Dzięki temu metoda ta daje lepsze przybliżenie rozwiązania niż klasyczna metoda Eulera.
$$y_{i+1} = y_i + \frac{1}{2}h[f(x_i,y_i)+f(x_i+h,y_i+hf(x_i,y_i))]$$
- **Metoda Rungego-Kutty** jest metodą dwuetapową, czwartego rzędu, co oznacza, że jej dokładność jest wyższa niż metod Eulera i Heuna. Metoda ta jest oparta na obliczeniu kilku przybliżonych nachyleń w punkcie $x_i$ oraz w odpowiednich punktach wewnętrznych przedziału $[x_i,x_{i+1}]$. Następnie te przybliżenia są ważąco sumowane w celu uzyskania ostatecznego wyniku.
\begin{gather*}
y_{i+1}=y_i+\frac{h}{6}(k_1+2k_2+2k_3+k_4)\\[10pt]
k_1 = f(t_i,y_i)\\[10pt]
k_2 = f(t_i+\frac{1}{2}h,y_i+\frac{1}{2}hk_1)\\[10pt]
k_3 = f(t_i+\frac{1}{2}h, y_i+\frac{1}{2}hk_2)\\[10pt]
k_4 = f(t_i+h,y_i+hk_3)
\end{gather*}

# Wyznaczanie błędów dla metod numerycznych
Błędy mogą mierzyć odchylenie danych przybliżonych od wartości dokładnej. Można to robić na różne sposoby. 

Na wykresach przedstawimy **maksymalny błąd bezwzględny** dla metod Eulera oraz Heuna dla różnych kroków czasowych $h$. Ponieważ nie mamy rozwiązania analitycznego modeli, jako wartość dokładną będziemy przyjmować tą wyznaczoną **metodą RK4** (aby łatwiej było skonstruować wykresy, również dla różnych kroków czasowych). Poniżej zamieszczamy wzór użyty do obliczenia błędów:

* **Maksymalny błąd bezwzględny**: $L^\infty = \sup\limits_k|y_k-\hat{y}_k|$, gdzie $y_k$ jest wartością dokładną, a $\hat{y}_k$ jest wartością przybliżoną.


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def funkcja(Zdyskretyzowany_czas,alfa=0.2,beta=0.6,delta=0.1,C=0.3,ro=0.1):
    h = Zdyskretyzowany_czas[1]-Zdyskretyzowany_czas[0]
    S_wynik = [50]
    I_wynik = [0]
    Z_wynik = [0]
    R_wynik = [0]
    for t in Zdyskretyzowany_czas:
        S = S_wynik[-1] + h*(-beta*S_wynik[-1]*Z_wynik[-1]-delta*S_wynik[-1])
        I = I_wynik[-1] + h*(beta*S_wynik[-1]*Z_wynik[-1]-(ro+delta)*I_wynik[-1])
        Z = Z_wynik[-1] + h*(ro*I_wynik[-1]-alfa*S_wynik[-1]*Z_wynik[-1]+C*R_wynik[-1])
        R = R_wynik[-1] + h*(delta*(S_wynik[-1]+I_wynik[-1])+alfa*S_wynik[-1]*Z_wynik[-1]-C*R_wynik[-1])
        S_wynik.append(S)
        I_wynik.append(I)
        Z_wynik.append(Z)
        R_wynik.append(R)
    czas = [0] + list(Zdyskretyzowany_czas)
    return czas,S_wynik,I_wynik,Z_wynik,R_wynik

def funkcja2(Zdyskretyzowany_czas,alfa=0.2,beta=0.6,delta=0.1,C=0.3,ro=0.1):
    h = Zdyskretyzowany_czas[1] - Zdyskretyzowany_czas[0]
    S_wynik = [50]
    I_wynik = [0]
    Z_wynik = [0]
    R_wynik = [0]
    for t in Zdyskretyzowany_czas:
        S_1 = S_wynik[-1] + h*(-beta*S_wynik[-1]*Z_wynik[-1] - delta*S_wynik[-1])
        I_1 = I_wynik[-1] + h*(beta * S_wynik[-1] * Z_wynik[-1] - (ro+delta)*I_wynik[-1])
        Z_1 = Z_wynik[-1] + h*(ro*I_wynik[-1] - alfa*S_wynik[-1]*Z_wynik[-1] + C*R_wynik[-1])
        R_1 = R_wynik[-1] + h*(delta*(S_wynik[-1]+I_wynik[-1]) + alfa*S_wynik[-1]*Z_wynik[-1] - C*R_wynik[-1])
        S = S_wynik[-1] + 0.5 * h*(-beta*S_wynik[-1]*Z_wynik[-1] - delta*S_wynik[-1] - beta*S_1*Z_1 - delta*S_1)
        I = I_wynik[-1] + 0.5 * h*(beta * S_wynik[-1] * Z_wynik[-1] - (ro+delta)*I_wynik[-1] + beta * S_1 * Z_1 - (ro+delta)*I_1)
        Z = Z_wynik[-1] + 0.5 * h*(ro*I_wynik[-1] - alfa*S_wynik[-1]*Z_wynik[-1] + C*R_wynik[-1] + ro*I_1 - alfa*S_1*Z_1 + C*R_1)
        R = R_wynik[-1] + 0.5 * h*(delta*(S_wynik[-1]+I_wynik[-1]) + alfa*S_wynik[-1]*Z_wynik[-1] - C*R_wynik[-1] + delta*(S_1+I_1) + alfa*S_1*Z_1 - C*R_1)
        S_wynik.append(S)
        I_wynik.append(I)
        Z_wynik.append(Z)
        R_wynik.append(R)
    czas = [0] + list(Zdyskretyzowany_czas)
    return czas,S_wynik,I_wynik,Z_wynik,R_wynik

def funkcja3(Zdyskretyzowany_czas,alfa=0.2,beta=0.6,delta=0.1,C=0.3,ro=0.1):
    h = Zdyskretyzowany_czas[1] - Zdyskretyzowany_czas[0]
    S_wynik = [50]
    I_wynik = [0]
    Z_wynik = [0]
    R_wynik = [0]
    for t in Zdyskretyzowany_czas:
        sk1 = -beta*S_wynik[-1]*Z_wynik[-1] - delta*S_wynik[-1]
        ik1 = beta*S_wynik[-1]*Z_wynik[-1] - (ro+delta)*I_wynik[-1]
        zk1 = ro * I_wynik[-1] - alfa*S_wynik[-1]*Z_wynik[-1] + C*R_wynik[-1]
        rk1 = delta*(S_wynik[-1]+I_wynik[-1]) + alfa*S_wynik[-1]*Z_wynik[-1] - C*R_wynik[-1]
        sk2 = -beta*(S_wynik[-1] + 0.5*h*sk1)*(Z_wynik[-1]+0.5*h*zk1) - delta*(S_wynik[-1] + 0.5*h*sk1)
        ik2 = beta*(S_wynik[-1] + 0.5*h*sk1)*(Z_wynik[-1]+0.5*h*zk1) - (ro+delta)*(I_wynik[-1]+0.5*h*ik1)
        zk2 = ro * (I_wynik[-1]+0.5*h*ik1) - alfa*(S_wynik[-1] + 0.5*h*sk1)*(Z_wynik[-1]+0.5*h*zk1) + C*(R_wynik[-1]+0.5*h*rk1)
        rk2 = delta*((S_wynik[-1] + 0.5*h*sk1)+(I_wynik[-1]+0.5*h*ik1)) + alfa*(S_wynik[-1] + 0.5*h*sk1)*(Z_wynik[-1]+0.5*h*zk1) - C*(R_wynik[-1]+0.5*h*rk1)
        sk3 = -beta*(S_wynik[-1] + 0.5*h*sk2)*(Z_wynik[-1]+0.5*h*zk2) - delta*(S_wynik[-1] + 0.5*h*sk2)
        ik3 = beta*(S_wynik[-1] + 0.5*h*sk2)*(Z_wynik[-1]+0.5*h*zk2) - (ro+delta)*(I_wynik[-1]+0.5*h*ik2)
        zk3 = ro * (I_wynik[-1]+0.5*h*ik2) - alfa*(S_wynik[-1] + 0.5*h*sk2)*(Z_wynik[-1]+0.5*h*zk2) + C*(R_wynik[-1]+0.5*h*rk2)
        rk3 = delta*((S_wynik[-1] + 0.5*h*sk2)+(I_wynik[-1]+0.5*h*ik2)) + alfa*(S_wynik[-1] + 0.5*h*sk2)*(Z_wynik[-1]+0.5*h*zk2) - C*(R_wynik[-1]+0.5*h*rk2)
        sk4 = -beta*(S_wynik[-1] + h*sk3)*(Z_wynik[-1]+ h*zk3) - delta*(S_wynik[-1] + h*sk3)
        ik4 = beta*(S_wynik[-1] + h*sk3)*(Z_wynik[-1]+ h*zk3) - (ro+delta)*(I_wynik[-1]+ h*ik3)
        zk4 = ro * (I_wynik[-1]+h*ik3) - alfa*(S_wynik[-1] + h*sk3)*(Z_wynik[-1]+ h*zk3) + C*(R_wynik[-1]+ h*rk3)
        rk4 = delta*((S_wynik[-1] + h*sk3)+(I_wynik[-1]+ h*ik3)) + alfa*(S_wynik[-1] + h*sk3)*(Z_wynik[-1]+ h*zk3) - C*(R_wynik[-1]+ h*rk3)
        S = S_wynik[-1] + (1/6)*h*(sk1+2*sk2 + 2*sk3 + sk4)
        I = I_wynik[-1] + (1/6)*h*(ik1+2*ik2 + 2*ik3 + ik4)
        Z = Z_wynik[-1] + (1/6)*h*(zk1+2*zk2 + 2*zk3 + zk4)
        R = R_wynik[-1] + (1/6)*h*(rk1+2*rk2 + 2*rk3 + rk4)
        S_wynik.append(S)
        I_wynik.append(I)
        Z_wynik.append(Z)
        R_wynik.append(R)
    czas = [0] + list(Zdyskretyzowany_czas)
    return czas,S_wynik,I_wynik,Z_wynik,R_wynik


bledy_inf_Euler = []
for h in np.linspace(0.001,0.01,100):
    Zdyskretyzowany_czas = np.linspace(h,20,int(20/h))
    doklwynik = funkcja3(Zdyskretyzowany_czas)[3]
    przyblizony_wynik = funkcja(Zdyskretyzowany_czas)[3]
    supremum = []
    for i in range(int(20/h)):
        blad = np.abs(doklwynik[i] - przyblizony_wynik[i])
        supremum.append(blad)
    supremum.sort()
    bledy_inf_Euler.append(supremum[-1])

bledy_inf_Heun = []
for h in np.linspace(0.001,0.01,100):
    Zdyskretyzowany_czas = np.linspace(h,20,int(20/h))
    doklwynik = funkcja3(Zdyskretyzowany_czas)[3]
    przyblizony_wynik = funkcja2(Zdyskretyzowany_czas)[3]
    supremum = []
    for i in range(int(20/h)):
        blad = np.abs(doklwynik[i] - przyblizony_wynik[i])
        supremum.append(blad)
    supremum.sort()
    bledy_inf_Heun.append(supremum[-1])

fig,axs = plt.subplots(1,2,figsize=(15,5))
axs[0].plot(np.linspace(0.001,0.01,100),bledy_inf_Euler)
axs[0].set_title('Maksymalny błąd bezwzględny - Euler')
axs[0].set_xlabel('krok czasowy')
axs[0].set_ylabel('błąd')
axs[0].grid()
axs[1].plot(np.linspace(0.001,0.01,100),bledy_inf_Heun)
axs[1].set_title('Maksymalny błąd bezwzględny - Heun')
axs[1].set_xlabel('krok czasowy')
axs[1].set_ylabel('błąd')
axs[1].grid()
```

# Porównanie obu modeli oraz wnioski
Porównując **model zombie z grupą zainfekowanych** oraz **model z możliwością wyleczenia**, można wyciągnąć następujące wnioski:

#### **Główne różnice w założeniach modeli**
* **Model z grupą zainfekowanych**: W tym modelu grupa zainfekowanych $I$ to osoby, które zostały zaatakowane przez zombie, ale jeszcze nie stały się zombie. Zainfekowani mogą albo przejść do grupy zombie $Z$, jeśli proces infekcji zakończy się, albo przejść do grupy zmarłych $R$, jeśli umrą przed pełną przemianą.
* **Model z możliwością wyleczenia**: Tutaj zainfekowani $I$ mogą albo zostać zombie $Z$, jeśli proces infekcji postępuje, albo zostać **wyleczeni**, co oznacza, że wracają do stanu zdrowia w przedstawionym w projekcie modelu nie są odporni na ponowne zarażenie. Wprowadzenie mechanizmu wyleczenia umożliwia aktywne zarządzanie epidemią i zmniejszanie liczby zainfekowanych, zanim staną się zombie.

#### **Wpływ na wyniszczenie populacji**
* **Model z grupą zainfekowanych**: Brak mechanizmu wyleczenia sprawia, że liczba zainfekowanych rośnie, aż do momentu, gdy populacja podatnych $S$ zostaje wyczerpana. Wynik końcowy jest katastrofalny: nastąpi dominacja zombie i wyniszczenie całej populacji żywych ludzi.
* **Model z możliwością wyleczenia**: Możliwość wyleczenia oznacza, że część zainfekowanych $I$ może zostać przywrócona do zdrowia, zmniejszając presję infekcji w układzie.Dzięki temu równowaga może zostać osiągnięta, gdzie populacja podatnych i wyleczonych współistnieje z mniejszą liczbą zainfekowanych  i zombie. Epidemia może zostać kontrolowana.

#### **Scenariusze końcowe**
* **Model z grupą zainfekowanych**: Scenariusze końcowe są bardziej pesymistyczne. Możliwe rezultaty to albo zombie całkowicie dominują, z niewielką lub zerową liczbą osób podatnych i zainfekowanych.
* **Model z możliwością wyleczenia**: Scenariusze końcowe są bardziej optymistyczne. Możliwe jest całkowite opanowanie epidemii, jeśli współczynnik wyleczenia $c$ przeważa nad współczynnikiem infekcji $\beta$. Przy odpowiednich parametrach układ osiąga stabilną równowagę, w której różne grupy $S$, $I$, $R$ i $Z$ współistnieją, bez wyniszczenia całej populacji.

#### **Znaczenie dla przewidywania dynamiki epidemii**
* **Model z grupą zainfekowanych** dobrze opisuje sytuacje, w których brak skutecznych środków zaradczych (np. pandemie bez szczepionek).
* **Model z możliwością wyleczenia** pozwala analizować scenariusze z aktywnymi interwencjami, takimi jak leczenie zainfekowanych lub zwiększanie odporności populacji. Jest bardziej realistyczny w kontekście współczesnych epidemii i strategii zdrowia publicznego.

#### **Wnioski**
Dodanie mechanizmu wyleczenia wprowadza bardziej optymistyczną perspektywę w porównaniu z modelem opartym wyłącznie na grupie zainfekowanych. Pozwala to na skuteczne zarządzanie epidemią i ochronę populacji, podczas gdy model z grupą zainfekowanych prowadzi do nieuniknionej katastrofy, chyba że zostaną zastosowane bardzo drastyczne środki (np. eliminacja zombie).

# Wykresy dla obu modeli z możliwością regulacji parametrów
Poniższe wykresy ilustrują przebieg dynamiki obu modeli: podstawowego modelu z grupą zainfekowanych oraz rozszerzonego modelu uwzględniającego możliwość wyleczenia.

Wykresy pozwalają użytkownikowi samodzielnie wprowadzić wartości parametrów modeli (takich jak współczynniki transmisji choroby, wyzdrowienia czy umieralności), a następnie zobaczyć, jak różnice w założeniach wpływają na przebieg dynamiki populacji zainfekowanej w czasie.

Takie podejście nie tylko umożliwia lepsze zrozumienie wpływu kluczowych parametrów na zachowanie układu, ale także daje czytelnikowi możliwość eksploracji wyników i analizy modeli w różnych warunkach.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interactive, HBox
def wykres_zainfekowani(alfa, beta, delta, ro, C,  S_0, I_0, Z_0, R_0):
    zdyskretyzowany_czas = np.linspace(0.001,20,19999)
    h = zdyskretyzowany_czas[1]-zdyskretyzowany_czas[0]
    S_wynik = np.zeros(len(zdyskretyzowany_czas))
    I_wynik = np.zeros(len(zdyskretyzowany_czas))
    Z_wynik = np.zeros(len(zdyskretyzowany_czas))
    R_wynik = np.zeros(len(zdyskretyzowany_czas))
    S_wynik[0] = S_0
    I_wynik[0] = I_0
    Z_wynik[0] = Z_0
    R_wynik[0] = R_0
    for t in range(1, len(zdyskretyzowany_czas)):
        S_wynik[t] = S_wynik[t-1] + h * (-beta * S_wynik[t-1] * Z_wynik[t-1] - delta * S_wynik[t-1])
        I_wynik[t] = I_wynik[t-1] + h * (beta * S_wynik[t-1] * Z_wynik[t-1] - (ro + delta) * I_wynik[t-1])
        Z_wynik[t] = Z_wynik[t-1] + h * (ro * I_wynik[t-1] - alfa * S_wynik[t-1] * Z_wynik[t-1] + C * R_wynik[t-1])
        R_wynik[t] = R_wynik[t-1] + h * (delta * (S_wynik[t-1] + I_wynik[t-1]) + alfa * S_wynik[t-1] * Z_wynik[t-1] - C * R_wynik[t-1])
    plt.plot(zdyskretyzowany_czas, S_wynik, label='Podatni')
    plt.plot(zdyskretyzowany_czas, I_wynik, label='Zainfekowani')
    plt.plot(zdyskretyzowany_czas, Z_wynik, label='Zombie')
    plt.plot(zdyskretyzowany_czas, R_wynik, label='zmarli')
    plt.title('Model z grupą zainfekowanych')
    plt.xlabel('Czas')
    plt.ylabel('Populacje')
    plt.legend()
    plt.grid(True)
    plt.show()
def wykres_wyleczenie(alfa, beta, delta, ro, c, C, S_0, I_0, Z_0, R_0):
    Zdyskretyzowany_czas = np.linspace(0.001,20,19999)
    h = Zdyskretyzowany_czas[1]-Zdyskretyzowany_czas[0]
    S_wynik = np.zeros(len(Zdyskretyzowany_czas))
    I_wynik = np.zeros(len(Zdyskretyzowany_czas))
    Z_wynik = np.zeros(len(Zdyskretyzowany_czas))
    R_wynik = np.zeros(len(Zdyskretyzowany_czas))
    S_wynik[0] = S_0
    I_wynik[0] = I_0
    Z_wynik[0] = Z_0
    R_wynik[0] = R_0
    for t in range(1, len(Zdyskretyzowany_czas)):
        S_wynik[t] = S_wynik[t-1] + h * (-beta * S_wynik[t-1] * Z_wynik[t-1] - delta * S_wynik[t-1] + c * Z_wynik[t-1])
        I_wynik[t] = I_wynik[t-1] + h * (beta * S_wynik[t-1] * Z_wynik[t-1] - (ro + delta) * I_wynik[t-1])
        Z_wynik[t] = Z_wynik[t-1] + h * (ro * I_wynik[t-1] - alfa * S_wynik[t-1] * Z_wynik[t-1] - c * Z_wynik[t-1] + C * R_wynik[t-1])
        R_wynik[t] = R_wynik[t-1] + h * (delta * (S_wynik[t-1] + I_wynik[t-1]) + alfa * S_wynik[t-1] * Z_wynik[t-1] - C * R_wynik[t-1])
    plt.plot(Zdyskretyzowany_czas, S_wynik, label='Podatni')
    plt.plot(Zdyskretyzowany_czas, I_wynik, label='Zainfekowani')
    plt.plot(Zdyskretyzowany_czas, Z_wynik, label='Zombie')
    plt.plot(Zdyskretyzowany_czas, R_wynik, label='zmarli')
    plt.title('Model z możliwością wyleczenia')
    plt.xlabel('Czas')
    plt.ylabel('Populacje')
    plt.legend()
    plt.grid(True)
    plt.show()
interaktywny_zainfekowani = interactive(
    wykres_zainfekowani,
    alfa=widgets.FloatSlider(value=0.2, min=0, max=1, step=0.01, description= 'α :'),
    beta=widgets.FloatSlider(value=0.6, min=0, max=1, step=0.01, description='β :'),
    delta=widgets.FloatSlider(value=0.1, min=0, max=1, step=0.01, description='δ :'),
    ro=widgets.FloatSlider(value=0.1, min=0, max=1, step=0.01, description='ρ :'),
    C=widgets.FloatSlider(value=0.3, min=0, max=1, step=0.01, description='ζ :'),
    S_0=widgets.FloatSlider(value=50, min=0, max=100, step=1, description='S(0):'),
    I_0=widgets.FloatSlider(value=0, min=0, max=100, step=1, description='I(0):'),
    Z_0=widgets.FloatSlider(value=0, min=0, max=100, step=1, description='Z(0):'),
    R_0=widgets.FloatSlider(value=0, min=0, max=100, step=1, description='R(0):'),
)
interaktywny_wyleczenie = interactive(
    wykres_wyleczenie,
    alfa=widgets.FloatSlider(value=0.2, min=0, max=1, step=0.01, description= 'α: '),
    beta=widgets.FloatSlider(value=0.6, min=0, max=1, step=0.01, description='β: '),
    delta=widgets.FloatSlider(value=0.1, min=0, max=1, step=0.01, description='δ :'),
    ro=widgets.FloatSlider(value=0.1, min=0, max=1, step=0.01, description='ρ :'),
    c=widgets.FloatSlider(value=0.1, min=0, max=1, step=0.01, description='c :'),
    C=widgets.FloatSlider(value=0.3, min=0, max=1, step=0.01, description='ζ :'),
    S_0=widgets.FloatSlider(value=50, min=0, max=100, step=1, description='S(0):'),
    I_0=widgets.FloatSlider(value=0, min=0, max=100, step=1, description='I(0):'),
    Z_0=widgets.FloatSlider(value=0, min=0, max=100, step=1, description='Z(0):'),
    R_0=widgets.FloatSlider(value=0, min=0, max=100, step=1, description='R(0):'),
)
display(HBox([interaktywny_zainfekowani, interaktywny_wyleczenie], layout=widgets.Layout(align_items='flex-end')))
```
